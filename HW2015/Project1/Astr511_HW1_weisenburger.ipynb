The data file Astr511HW1data.dat (linked to class webpage as a gzipped file) contains SDSS measurements for about 600,000 stars with b > 80deg (i.e. within 10deg from the north galactic pole) and 14 < r < 21. The data are listed as one line per star, with each line containing the following quantities:
* **ra dec**: right ascension and declination (J2000.0) in decimal degrees
* **run**: SDSS observing night identifier
* **Ar**: the value of the r band ISM extinction used to correct photometry (adopted from the SFD maps; for bands other than r standard SDSS coefficients are used)
* **u g r i z**: SDSS photometry (corrected for the ISM extinction)
* **uErr gErr rErr iErr zErr**: photometric errors
* **pmL pmB**: proper motion vector components in the longitudinal and latitudinal directions (mas/yr); set to 999.99 when no measurement is available
* **pmErr**: mean proper motion error (mas/yr); set to 999.99 when no measurement is available
---
For stars from this file, we compute absolute magnitude using a photometric parallax relation, $M_r (g − i, [F e/H ])$, given by eqs. A2, A3 and A7 from Ivezic et al. 2008 [(ApJ, 684, 287)](http://adsabs.harvard.edu/abs/2008ApJ...684..287I). For computing metallicity, $[Fe/H]$, instead of their eq. 4, we use an updated expression from Bond et al. 2010 [(ApJ, 716, 1)](http://adsabs.harvard.edu/cgi-bin/bib_query?arXiv:0909.0013). As this expression is valid only for $g − r$ < 0.6, we use $[Fe/H]$ = −0.6 for redder stars.

```python
%matplotlib inline  
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import math
from pylab import *
```

```python
(ra, dec, run, ar, umag, gmag, rmag, imag, zmag, uerr, 
 gerr, rerr, ierr, zerr, pml, pmb, pmerr) = np.loadtxt('Astr511HW1data.dat', unpack=True)

# color indices
x = umag - gmag
y = gmag - rmag

# A-J coefficients
ac, bc, cc, dc, ec, fc, gc, hc, ic, jc = (-13.13,
                                          14.09, 28.04, -5.51, -5.90, -58.68, 
                                          9.14, -20.61, 0.0, 58.20)
```

```python
# Fe/H from Bond 10
feh = (ac + (bc*x) + (cc*y) + (dc*x*y) + (ec*(x**2)) + (fc*(y**2)) + 
       (gc*(x**2)*y) + (hc*x*(y**2)) + (ic*(x**3)) + (jc*(y**3)))

# Fe/H eqn doesn't apply to redder stars
red = y > .6
feh[red] = -0.6
```

```python
# A2 from Ivezic 08
deltaM = 4.50 - (1.11*feh) - (0.18*(feh**2))

# A7 from  Ivezic 08
z = gmag - imag
Mr0 = -5.06 + (14.32*z) - (12.97*(z**2)) + (6.127*(z**3)) - (1.267*(z**4)) + (0.0967*(z**5))

# A3 from Ivezic 08
Mr = Mr0 + deltaM
# distances in kpc
dist = (10**(((rmag - Mr)/5.)+1) ) /1000.
```

### 1. Ln($\ \rho/\rho_{0}$\) vs. $Z$ for stars with 0.2 < $g-r$ < 0.4

Since b > 80$^\circ$, for these stars the distance from the galactic plane, $Z$, and the distance from us, $D$, are approximately the same. Using $Z = D$, where $D$ is computed from
$r − Mr$ = 5 ∗ log ($D$/(10pc)), we can also calculate the stellar number density, ln($\rho$), where $\rho(Z) = N(Z)/V(Z)$ and $V(Z)$ is the bin volume in kpc$^3$.

$$ dV = r^2\ sin\ \phi\ d\theta\  d\phi\  dr $$

$$ V = \int_{r_1}^{r_2} r^2 \int_0^{10^\circ} d\phi \int_0^{2\pi} d\theta $$

$$ V = \frac{{r_2}^3 - {r_1}^3}{3}\ 2\pi\ (1-\cos{10}^\circ) $$

```python
# function to compute ln(rho) given min/max color indices, 
# number of bins (default=50) and metallicity criteria (default=no metal cut)

def density(gr_min, gr_max, nbin=50, metal='none'):
    
    #  first get distances of stars that satisfy color/metal cut
    if metal.lower() == 'high':
        mask = (y > gr_min) & (y < gr_max) & (feh > -1.0)
    elif metal.lower() == 'low':
        mask = (y > gr_min) & (y < gr_max) & (feh < -1.0)        
    else:
        mask = (y > gr_min) & (y < gr_max)
        
    z = dist[mask]
    darr = np.linspace(0, np.max(dist), nbin)

    # compute volume of each evenly spaced bin of Z
    vol = (((darr[1:])**3 - (darr[:-1])**3)/3.) * (2*math.pi) * (1 - math.cos(math.radians(10)))

    # count number of stars within each volume
    tot = np.zeros(nbin)
    for x in range(nbin-1):
        tot[x] = sum( (z < darr[x+1]) & (z > darr[x]))

    # compute density (stars/volume)
    rho = tot[:-1]/vol
    #print rho
    lnrho = [np.log(n/rho[0]) for n in rho]

    # return Z, N, ln(rho) and rmag arrays
    return (darr[:-1]), tot[:-1], lnrho, rmag[mask]
```

```python
# compute ln(rho) for 0.2 < g-r < 0.4
z1, n1, rho1, rmag1 = density(0.2, 0.4)
```

```python
plt.figure(figsize=(10,6))
plt.plot(z1, rho1, ls='steps', lw=3)

plt.tick_params(axis='both', which='major', labelsize=16)
plt.xlabel('$Z$ (kpc)', fontsize=22)
plt.ylabel(r'ln($\rho/\rho_{0}$)', fontsize=22)
ylim(-20,0)

# fit exponential?
coefficients = np.polyfit(z1[2:15], rho1[2:15], 1)
polynomial = np.poly1d(coefficients)
ys = polynomial(np.linspace(0,80,1000))
plt.plot(np.linspace(0,80,1000), ys, 'r', lw=2, alpha=.7)

# use poisson counting statistic to find where uncertainty exceeds 5%
pois = 1/np.sqrt(n1)
uncert = min(enumerate(pois), key=lambda x: abs(x[1]-0.5))
plt.vlines(uncert[0], -25, 10, lw=3, linestyles='dashed', alpha=.5)
plt.text(35, -5, 'Fractional uncertainty > 5%\nat $Z\sim 31$ kpc', fontsize=17)
```

The blue line shows the normalized density distribution of stars with $0.2 < g-r < 0.4$ as a function of Galactic height. To estimate to what distance we believe this distribution, we can compute the fractional uncertainty of each bin. The gray dashed line shows the distance limit where this uncertainty exceeds 5%. This cut is an arbitrary one, but we can convince ourselves that this is a good estimate given additional information:

- Ivezic et al. 2012 [(ARAA, 50 251)](http://adsabs.harvard.edu/abs/2012ARA%26A..50..251I) summarizes previous studies of the Galactic halo and the exponential falloff of stars as a function of Galactic height. They report that beyond 25 kpc, this falloff only grows steeper; therefore, that we begin to see an *increase* in stellar density is likely a false alarm.

- In a similar analysis, Juric et al. 2008 [(ApJ, 673, 864)](http://adsabs.harvard.edu/abs/2008ApJ...673..864J) used only stars with $Z\lt20$ kpc, suggesting that my 31 kpc cut may even be too optimistic. Additionally, Juric 2008 estimates that for a star with $0.2 < g-r < 0.4$ and a distance of $Z\sim3$ kpc, it has a $\sim$20% distance uncertainty and errors increase with distance.

This figure shows the normalized density distributions of stars as a function of Galactic height and $g-r$ color index.

Redder bins (larger $g-r$ indices) truncate at closer distances because SDSS could not probe fainter, lower mass stars at large distances due to its $r\sim21$ limit. The bluest bin extends to much larger distances because these stars are intrinsically brighter. With smaller number statistics and larger uncertainties at faint magnitudes, we expect larger systematic errors in the redder bins.

It's important to note that the redder bins will also have contamination by evolved giants with underestimated distances, adding to the uncertainty of those bins.

The steeper fall off of redder bins also demonstrates Malmquist bias. For the bluer bins with brighter stars, we see high density distributions, but the redder bins show a more drastic falloff with distance, i.e. intrinsically faint objects are not observed due to selection effects.

### 5. Faint $r$ band limit to map the ln($\ \rho$\) vs. $Z$ profile to 100 kpc

Here we can extrapolate our SDSS sample obtain and estimate of the $r$ limit needed to probe stars at 100 kpc:

```python
from astroML.stats import binned_statistic_2d
import cubehelix
mask = (dist < 31) & (y > 0.2) & (y < 0.4)

xx = rmag[mask]  #[:100000]
yy = np.log10(dist[mask])  #[:100000])

plt.figure(figsize=(10,6))

N, xedges, yedges = binned_statistic_2d(xx, yy, yy,
                                        'count', bins=100)
#cmap = plt.cm.jet
cmap = cubehelix.cmap(reverse=True, start=0.3, rot=-0.5)
cmap.set_bad('w', 1.)

cmap_multicolor = plt.cm.jet
cmap_multicolor.set_bad('w', 1.)

# plot density color contours
plt.imshow(np.log10(N.T), origin='lower',
           extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
           aspect='auto', interpolation='nearest', cmap=cmap)

# fit linear regression to data
coefficients = np.polyfit(xx, yy, 1)
polynomial = np.poly1d(coefficients)
ys = polynomial(np.linspace(12,28,100))
plt.plot(np.linspace(12,28,100), ys, 'r', lw=2)
plt.text(13.6, .5, 'Linear fit', color='r', fontsize=17, rotation=30)

plt.ylim(-0.5, 2.5)
plt.vlines(24.75, -1.5, 3, 'k', lw=2)
plt.text(25, 1.05, r'$r$ limit', fontsize=17)
plt.xlim(13,28)
plt.hlines(2, 12, 28, 'k', lw=2)
plt.text(18, 2.05, '100 kpc', fontsize=17)

plt.xlabel(r'$r$ (mag)', fontsize=22)
plt.ylabel(r'log($Z$)', fontsize=22)
```

At $r\sim24.8$ we could observe F stars at 100 kpc.

Even if this is a rough estimate, Galactic stratigraphy dictates that older stars are more likely to be found at higher Galactic latitudes -- these stars have longer lifetimes during which they have interacted with nearby stars and been sent into orbits farther from the plane (dynamical heating). So at 100 kpc out of the Plane, there are more stars with longer lifetimes, i.e. M dwarfs. Could we observe those? Can we with LSST?
